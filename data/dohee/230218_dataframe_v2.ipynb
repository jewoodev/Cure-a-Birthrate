```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
unsold_house = pd.read_excel('./2007-2022_미분양현황.xlsx', index_col = 0)
apt_rent = pd.read_excel('./2003-2022_아파트전세_가격지수.xlsx', index_col = 0)
birth = pd.read_excel('./1997-2021_총_출생아수.xlsx', index_col = 0)
```

## 전국 데이터프레임

```python
unsold_house.loc['2007-01-01' : '2021-12-01'].index
```

```python
# 빈 데이터 프레임 형성
merge_v1 = pd.DataFrame(np.nan, index = unsold_house.loc['2007-01-01' : '2021-12-01'].index, columns = ['unsold_house', 'apt_rent', 'birth(y)'])

# 2007~2021 행 중 '전국' 값 대입하기
merge_v1['unsold_house'] = unsold_house.loc['2007-01-01' : '2021-12-01']['전국'].values
merge_v1['apt_rent'] = apt_rent.loc['2007-01-01' : '2021-12-01']['전국'].values
merge_v1['birth(y)'] = birth.loc['2007-01-01' : '2021-12-01']['전국'].values
merge_v1
```

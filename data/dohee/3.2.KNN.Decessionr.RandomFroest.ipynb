```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

path = 'C:/playdata/To_raise_a_child_well/data/dohee/'
sum_df = pd.read_excel(path+'전국데이터프레임2006_2021.xlsx', index_col = 0)
sum_df
```

```python
from sklearn.model_selection import train_test_split
x = sum_df.drop(['출산아수/가임기여성의수'], axis = 1).loc['2006-01-01' : '2021-12-01'].values
y = sum_df['출산아수/가임기여성의수'].loc['2006-01-01' : '2021-12-01']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
x_train.shape, x_test.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# 데이터 분리
from sklearn.model_selection import train_test_split
x = sum_df.drop(['출산아수/가임기여성의수'], axis = 1).loc['2006-01-01' : '2021-12-01'].values
y = sum_df['출산아수/가임기여성의수'].loc['2006-01-01' : '2021-12-01']
            
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
x_train.shape, x_test.shape

# 데이터 프레임
knn_reg_df = pd.DataFrame(index = ['mse', 'rmse', 'r2'])
dt_reg_df = pd.DataFrame(index = ['mse', 'rmse', 'r2'])
rf_reg_df = pd.DataFrame(index = ['mse', 'rmse', 'r2'])

# max_depth, n_neighbor 파라미터 : i
for i in range(1, 6) : 
    # n_neighbor 파라미터 : j
    for j in range(10, 101, 10) :
        knn_reg = KNeighborsRegressor(n_neighbors = i)
        dt_reg = DecisionTreeRegressor(max_depth = i)
        rf_reg = RandomForestRegressor(max_depth = i, n_estimators = j)

        # 학습        
        knn_reg.fit(x_train, y_train)
        dt_reg.fit(x_train, y_train)
        rf_reg.fit(x_train, y_train)
        
        # 예측
        y_pred_knn = knn_reg.predict(x_test)
        y_pred_dt = dt_reg.predict(x_test)
        y_pred_rf = rf_reg.predict(x_test)
        
        # 모델 평가 및 데이터프레임, r2 최대값 출력
        from sklearn.metrics import mean_squared_error, r2_score
        # knn
        mse = mean_squared_error(y_test, y_pred_knn)  
        rmse = np.sqrt(mse)                       
        r2 = r2_score(y_test, y_pred_knn)    
        total = np.round((mse, rmse, r2), 3)
        
        knn_reg_df['n_neighbors='+str(i)] = total
        knn_reg_df
        
        # dt
        mse = mean_squared_error(y_test, y_pred_dt)  
        rmse = np.sqrt(mse)                       
        r2 = r2_score(y_test, y_pred_dt)    
        total = np.round((mse, rmse, r2), 3)
        
        dt_reg_df['max_depth='+str(i)] = total
        dt_reg_df
        
        # rf
        mse = mean_squared_error(y_test, y_pred_rf)  
        rmse = np.sqrt(mse)                       
        r2 = r2_score(y_test, y_pred_rf)    
        total = np.round((mse, rmse, r2), 3)
        
        rf_reg_df['max_depth='+str(i)+'n_estimators='+str(j)] = total
        rf_reg_df
```

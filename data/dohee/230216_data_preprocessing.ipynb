#### 영유아 1인당 보육시설 수 : 보육시설 수 / 0-6세의 총 영유아 인구

```python
# 1) 보육시설 수 (childcare facilities)
childcare_fc = pd.read_excel('./0_데이터 전처리 전/2008-2021_영유아, 보육시설 종합본.xlsx', sheet_name = '보육시설 수')

# 데이터프레임 설정
# index를 행정구역으로 설정
childcare_fc = childcare_fc.set_index('보육시설 수')

# 열 재지정
childcare_fc.columns = ['2008년', '2009년', '2010년', '2011년', '2012년', '2013년', '2014년', '2015년', '2016년', 
                                       '2017년', '2018년', '2019년', '2020년', '2021년']
childcare_fc = childcare_fc.drop('행정기관')
childcare_fc
```

```python
# 2) 영유아 수 (infants)
infants = pd.read_excel('./0_데이터 전처리 전/2008-2021_영유아, 보육시설 종합본.xlsx', sheet_name = '영유아 수')

# 데이터프레임 설정
# index를 행정구역으로 설정
infants = infants.set_index('만 0 ~ 6세 영유아 수')

# 열 재지정 (2006 ~ 2007년도 추가)
infants.columns = ['2008년', '2009년', '2010년', '2011년', '2012년', '2013년', '2014년', '2015년', '2016년', 
                                       '2017년', '2018년', '2019년', '2020년', '2021년']
infants = infants.drop('행정기관')
infants
```

```python
infants['2008년'][1] / childcare_fc['2008년'][1]
```

```python
# 3) 영유아 1인당 보육시설 수 (childcare_fc_per_one_infants)

## [1] 빈 데이터프레임 형성
infants.shape, childcare_fc.shape
index = ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', 
         '세종특별자치시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']

childcare_fc_per_one_infants = pd.DataFrame(np.nan, index = index, columns = infants.columns)
# 원하는 위치에 열 추가
childcare_fc_per_one_infants.insert(0, '2006년', np.nan)
childcare_fc_per_one_infants.insert(1, '2007년', np.nan)
childcare_fc_per_one_infants
```

```python
## [2] 나눗셈 연산을 위한 보육시설 수와 영유아수의 0값 --> 1로 변경
childcare_fc.iloc[8] = childcare_fc.iloc[8].replace(0, 1)
infants.iloc[8] = infants.iloc[8].replace(0, 1)

# [3] 영유아 1인당 보육시설 수
for col in list(infants.columns) :
    for idx in range(len(infants.index)) : 
        childcare_fc_per_one_infants[col][idx] = childcare_fc[col][idx] / infants[col][idx]   #infants['2008년'][1]
        childcare_fc_per_one_infants = round(childcare_fc_per_one_infants, 4)
        
        # 세종시 1값 --> 0으로 변경
        childcare_fc_per_one_infants.iloc[8] = childcare_fc_per_one_infants.iloc[8].replace(1, 0)
        
# 2006년, 2007년 nan --> 0으로 변경
childcare_fc_per_one_infants.fillna(0, inplace = True)
childcare_fc_per_one_infants
```

```python
childcare_fc_per_one_infants.to_excel('2006-2021_영유아(만 0-6세) 천명당 보육시설 수.xlsx')
```

#### 종합체육시설 수

```python
# 체육시설(sports_fc1)
sports_fc1 = pd.read_excel('./0_데이터 전처리 전/2006-2021_종합체육시설.xlsx', sheet_name = '체육시설')
sports_fc1 = sports_fc1.drop([0, 1])
sports_fc1 = sports_fc1.set_index('소재지(시도)별(1)')
sports_fc1.index.name = None
sports_fc1.fillna(0, inplace = True)

# 운동장(sports_fc2)
sports_fc2 = pd.read_excel('./0_데이터 전처리 전/2006-2021_종합체육시설.xlsx', sheet_name = '운동장')
sports_fc2 = sports_fc2.drop([0, 1])
sports_fc2 = sports_fc2.set_index('소재지(시도)별(1)')
sports_fc2.index.name = None
sports_fc2.fillna(0, inplace = True)
# 2018 ~ 2021 열 추가
sports_fc2['2018'] = 0
sports_fc2['2019'] = 0
sports_fc2['2020'] = 0
sports_fc2['2021'] = 0

# 운동장 + 체육시설
total_sports_fc = sports_fc1 + sports_fc2
total_sports_fc
```

```python
total_sports_fc.to_excel('2006-2021_종합체육시설 수.xlsx')
```

#### 도시화율 : (지역의 도시면적 / 총 토지 면적) * 100

```python
# 총 토지 면적(land_size)
land_size = pd.read_excel('./0_데이터 전처리 전/2006-2021_총_토지면적(지목별).xlsx')
land_size.drop([0], inplace = True)
land_size.set_index('소재지(시군구)별(1)', inplace = True)
land_size.index.name = None
# 나눗셈을 위해 0값 --> 1로 변경
land_size.iloc[8] = land_size.iloc[8].replace(0, 1)
land_size

# 도시면적(city_size)
city_size = pd.read_excel('./0_데이터 전처리 전/2006-2021_도시지역면적.xlsx')
city_size.set_index('소재지(시군구)별', inplace = True)
city_size.index.name = None
# 나눗셈을 위해 0값 --> 1로 변경
city_size.iloc[8] = city_size.iloc[8].replace(0, 1)
city_size

# 도시화율
for idx in range(len(city_size.index)) : 
    city_size.iloc[idx] = (city_size.values / land_size.values)[idx]
    urbanization_rate = round(city_size, 4)
urbanization_rate.to_excel('2006-2021_도시화율.xlsx')
```

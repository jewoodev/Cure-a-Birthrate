```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Labor_Force_Participation_Rate.csv', encoding='CP949')
```

```python
# df.columns = df.iloc[0]
df = df.drop(df.index[0], axis=0) # 데이터 정리 완료
```

```python
list_to_append = []

for i in range(0, 284):
    list_to_append.append(df.index[i])
```

```python
for i in range(0, 284):
    list_to_append[i] = list_to_append[i].replace('.', '-')
```

```python
from datetime import datetime

for i in range(0, 284):
    list_to_append[i] = datetime.strptime(list_to_append[i], '%Y-%m')
```

## 년도별로 정리된 데이터를 추가작업, 지역별로 정리하기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('After_preprocessing/Total_bought_house.xlsx')
df
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.at[0,'Unnamed: 0'] = '서울22년'
df.rename(columns={'Unnamed: 0':'지역'},inplace=True)
```

```python
df['지역'] = df['지역'].apply(lambda x: x.split('년')[0])
# 0~10, 11~21, 22~32, 33~43...
```

```python
df['시점'] = df['지역'].apply(lambda x: '20'+x[2:] if len(x)==4 else('20'+x[3:] if len(x)==5 else '20'+x[4:]))
df['시점'] = df['시점'].astype('int')
```

```python
# 서울
seoul_df = df.loc[0:10,['시점','주택매매전체','아파트매매']]
seoul_df =seoul_df.sort_values(by='시점')
seoul_df['지역'] = 'Seoul'
seoul_df.set_index('시점',inplace=True)
seoul_df
```

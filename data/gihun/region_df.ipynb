```python
# 라이브러리
import numpy as np
import pandas as pd
import datetime as dt
```

```python
path = "dataset/"
dohee_path = "../dohee/"
jewoo_path = "../jewoo/After_preprocessing"
```

```python
# 파일불러오기
elder_df = pd.read_excel(path+'고령화인구비율최종.xlsx',index_col=0)
Divorced_df= pd.read_excel(path+'이혼건수최종.xlsx',index_col=0)
Unemp_df=pd.read_excel(path+'실업률최종.xlsx',index_col=0)
Married_df=pd.read_excel(path+'혼인건수최종.xlsx',index_col=0)
CPI_df=pd.read_excel(path+'소비자물가지수최종.xlsx',index_col=0)
girl=pd.read_excel(path+'행정구역여성인구수최종.xlsx',index_col=0)
number_of_birth = pd.read_excel(dohee_path+'1997-2021_총_출생아수.xlsx',index_col=0)
apart_price = pd.read_excel(dohee_path+'2003-2022_아파트전세_가격지수.xlsx',index_col=0)
unsold_df = pd.read_excel(dohee_path+'2003-2022_아파트전세_가격지수.xlsx',index_col=0)
bank_deposit_df = pd.read_excel(jewoo_path+'/Deposit_of_the_bank_of_deposit.xlsx',index_col=0)
fn_idp_df = pd.read_excel(jewoo_path+'/Labor_Force_Participation_Rate.xlsx',index_col=0) # 계, 제주도
fluct_df = pd.read_excel(jewoo_path+'/fluctuation_rate_of_land_price_v2.xlsx',index_col=0)
```

```python
# 1997~2021합치기
girl_df = girl.loc[:'2021-12-01']
number_of_birth_df = number_of_birth.loc['2008-01-01':'2021-12-01']
number_of_birth_df
```

```python
# 총 출산아 수 /15~49세 여성의 수
birth_rate_df = girl_df.copy()
birth_rate_df = number_of_birth_df/birth_rate_df
# 경제참가전처리
fn_idp_df.rename(columns={'계':'전국','제주도':'제주특별자치도'},inplace=True)
fn_idp_df.index.name = '시점'
fn_idp_df.reset_index(inplace=True)
fn_idp_df['시점'] = fn_idp_df['시점'].astype('str')
fn_idp_df['시점'] = fn_idp_df['시점'].apply(lambda x: x+'0' if len(x)==6 else x)
fn_idp_df.set_index('시점',inplace=True)
fn_idp_df.index = pd.to_datetime(fn_idp_df.index, format='%Y-%m')
```

```python
bank_deposit_df.replace('-',0,inplace=True)
bank_deposit_df = bank_deposit_df.apply(pd.to_numeric)
bank_deposit_df.info()
```

```python
# 파일 리스트
file_list = [Unemp_df, Divorced_df, elder_df, Married_df, CPI_df, birth_rate_df, apart_price, unsold_df, bank_deposit_df,fn_idp_df ,fluct_df]
```

```python
# 지역 데이터프레임 함수 
def get_resion_df(region_df, file_list, region):
    # 파일 리스트에서 각 지역 데이터 가져오기
    k = 0
    for i in file_list:
        k = k + 1
        region_df[k]= i[region]
    region_df.columns = ['실업률','이혼율','고령화비율','혼인건수','소비자물가지수','출산아수/가임기여성의수','아파트전세지수','미분양주택지수','은행예금지수','경제활동참가율','지가변동률']
    region_df['이혼율'] = region_df['이혼율'].values / region_df['혼인건수']
    return region_df
```

```python
# 각 지역 데이터프레임만들기
seoul_df = pd.DataFrame()
busan_df = pd.DataFrame()
daegu_df = pd.DataFrame()
incheon_df = pd.DataFrame()
gwangju_df = pd.DataFrame()
daejun_df = pd.DataFrame()
Ulsan_df = pd.DataFrame()
Sejong_df = pd.DataFrame()
# 경기도 데이터프레임
Gyeongi_df = pd.DataFrame()
# 강원도 데이터프레임
Gangwon_df = pd.DataFrame()
# 충청북도 데이터프레임
Nc_df = pd.DataFrame()
# 충청남도 데이터프레임
Sc_df = pd.DataFrame()
# 전라북도 데이터프레임
Nj_df = pd.DataFrame()
# 전라남도 데이터프레임
Sj_df = pd.DataFrame()
# 경상북도 데이터프레임
Ng_df = pd.DataFrame()
# 경상남도 데이터프레임
Sg_df = pd.DataFrame()
# 제주도 데이터프레임
Jeju_df = pd.DataFrame()
# 지역 데이터프레임 리스트
region_df_list = [seoul_df, busan_df, daegu_df, incheon_df, gwangju_df, daejun_df, Ulsan_df, Sejong_df, Gyeongi_df, Gangwon_df, Nc_df, Sc_df, Nj_df, Sj_df, Ng_df, Sg_df, Jeju_df]
# 각 시 컬럼 리스트
column_list = number_of_birth_df.columns
# column_list의 인덱스
idx = 1
for i in region_df_list:
    get_resion_df(i, file_list=file_list, region=column_list[idx])
    idx = idx +1
```

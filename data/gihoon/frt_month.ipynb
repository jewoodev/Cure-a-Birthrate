```python
# 고령인구비율 (고령인구/전체인구)파일불러오기
elder_df = pd.read_excel(path+'고령인구비율_시도_2008-2023.xlsx',index_col=0)
elder_df.drop('시점',axis=0,inplace=True)
elder_df
```

```python
# 자료형변환
elder_df.index= pd.to_datetime(elder_df.index, format="%Y-%m")
elder_df['세종특별자치시'].replace('-',0,inplace=True)
elder_df = elder_df.apply(pd.to_numeric)
elder_df
```

```python
# 이혼건수 불러오기
Divorced_df = pd.read_excel(path+'이혼건수_시도_1997-2021.xlsx')
```

```python
# 시점 열의 값이 float으로 되어있어서 float을 str로 바꿔준다음 yyyy.1만 yyyy.10으로 바꿔주기
Divorced_df['시점'] = Divorced_df['시점'].astype('str')
Divorced_df['시점'] = Divorced_df['시점'].apply(lambda x: x+'0' if len(x)==6 else x)
Divorced_df.set_index('시점',inplace=True)
```

```python
# 자료형변환
Divorced_df.index= pd.to_datetime(Divorced_df.index, format='%Y-%m')
Divorced_df['세종특별자치시'].replace('-',0,inplace=True)
Divorced_df = Divorced_df.apply(pd.to_numeric)
Divorced_df
```

```python
# 혼인건수 불러오기
Married_df = pd.read_excel(path+'혼인건수_시도_1997-2021.xlsx')
```

```python
# 시점 열의 값이 float으로 되어있어서 float을 str로 바꿔준다음 yyyy.1만 yyyy.10으로 바꿔주기
Married_df['시점'] = Married_df['시점'].astype('str')
Married_df['시점'] = Married_df['시점'].apply(lambda x: x+'0' if len(x)==6 else x)
Married_df.set_index('시점',inplace=True)
```

```python
# 자료형변환
Married_df.index= pd.to_datetime(Married_df.index, format='%Y-%m')
Married_df['세종특별자치시'].replace('-',0,inplace=True)
Married_df = Married_df.apply(pd.to_numeric)
Married_df
```

```python
# 소비자물가지수 자료 불러오기
CPI_df = pd.read_excel(path+'소비자물가지수_시도.xlsx',index_col=0)
CPI_df
```

```python
# 컬럼명 재설정
CPI_df.columns = Married_df.columns
# 1965~1989까지 결측치가 너무 많아서 제거
CPI_df = CPI_df.iloc[290:] # 한번만 할것
# 자료형변환
CPI_df.index= pd.to_datetime(CPI_df.index, format='%Y-%m')
CPI_df['세종특별자치시'].replace('-',0,inplace=True)
CPI_df = CPI_df.apply(pd.to_numeric)
```

```python
# 실업률 불러오기
Unemp_df = pd.read_excel(path+'행정구역_시도__성별_실업률_20230218162056.xlsx',index_col=0)
```

```python
# 자료컬럼행변환
Unemp_df = Unemp_df.T
# 성별행탈락
Unemp_df.drop('성별',axis=0, inplace=True)
# 세종시 결측치 0채우기
Unemp_df['세종특별자치시'].replace('-',0,inplace=True)
# 컬럼명변경
Unemp_df.rename(columns={'계':'전국','제주도':'제주특별자치시'},inplace=True)
# 인덱스 설정
Unemp_df.index.name = '시점'
Unemp_df.index= pd.to_datetime(Unemp_df.index, format='%Y-%m')
Unemp_df = Unemp_df.apply(pd.to_numeric)
Unemp_df
```
